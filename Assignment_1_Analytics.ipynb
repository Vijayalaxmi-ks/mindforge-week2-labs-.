```python
import pandas as pd 
import numpy as np 
# Simulating 50,000 retail transactions 
np.random.seed(42) 
data = { 
    "Transaction_ID": range(10000, 60000), 
    "Store_City": np.random.choice(["Mumbai", "Pune", "Delhi", "Bangalore"], 50000), 
    "Category": np.random.choice(["Electronics", "Clothing", "Groceries"], 50000), 
    "Revenue": np.random.normal(500, 150, 50000) 
} 
df = pd.DataFrame(data) 
# Injecting corruption (missing values) 
df.loc[df.sample(frac=0.05).index, "Revenue"] = np.nan 
df.head()
```

```python
# 1. Count how many Revenue values are missing
missing_count = df['Revenue'].isna().sum()
print(f"Number of missing Revenue values: {missing_count}")

# 2. Calculate the Median revenue
median_revenue = df['Revenue'].median()
print(f"Calculated Median Revenue: {median_revenue:.2f}")

# 3. Impute (fill) those missing values using the Median
df['Revenue'] = df['Revenue'].fillna(median_revenue)

# 4. Verification check to ensure 0 missing values remain
print(f"Missing Revenue values after cleanup: {df['Revenue'].isna().sum()}")
```

```python
# Calculate the Total (Sum) Revenue generated by each Store_City
city_revenue = df.groupby('Store_City')['Revenue'].sum()

print("--- Total Revenue by City ---")
print(city_revenue.map('${:,.2f}'.format)) # Formats numbers beautifully as currency
```
