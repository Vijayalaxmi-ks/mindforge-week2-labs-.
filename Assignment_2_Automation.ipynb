```python
server_logs = """ 
[INFO] 192.168.1.1 - User Logged In 
[ERROR] 10.0.0.5 - Database Connection Failed 
[INFO] 192.168.1.2 - Page Loaded 
[CRITICAL] 172.16.0.4 - Server Overheating! 
[ERROR] 10.0.0.9 - Password Invalid 
"""
```

```python
# Initialize the clean list for critical issues
critical_issues = []

# Split the server_logs string into a list of individual lines
log_lines = server_logs.strip().split('\n')

# Using a for loop to read every line
for line in log_lines:
    
    if "[ERROR]" in line or "[CRITICAL]" in line:
        
        # Extracting the IP address and the warning message
        # We split the line by the " - " divider to separate metadata from the message
        parts = line.split(" - ")
        message = parts[1] # "Database Connection Failed"
        
        # We split the first part by a space to separate the tag from the IP
        meta_parts = parts[0].split(" ")
        ip_address = meta_parts[1] # "10.0.0.5"
        
        # Append the extracted details as a clean string to our list
        critical_issues.append(f"IP: {ip_address} | Issue: {message}")
```

```python
# Output Generation
print("--- Critical Issues Detected ---")
for issue in critical_issues:
    print(issue)
```
